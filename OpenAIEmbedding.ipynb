```python
from openai import OpenAI
import pandas as pd
import numpy as np
import json
import time
```

```python
client = OpenAI(api_key="")

# =========================
# PATH
# =========================
INPUT_PATH = r"C:\Users\Yohanes\Documents\Lab\IntentMining-python-3.11\dataset_anonim_phase7_2024.xlsx"
OUTPUT_JSON = r"C:\Users\Yohanes\Documents\Lab\IntentMining-python-3.11\OpenAIEmbeddings_512.json"

# =========================
# NORMALIZATION
# =========================
def normalize_l2(x):
    x = np.array(x)
    norm = np.linalg.norm(x)
    if norm == 0:
        return x.tolist()
    return (x / norm).tolist()

# =========================
# LOAD XLSX
# =========================
df = pd.read_excel(INPUT_PATH)

# =========================
# MAIN LOOP
# =========================
results = []
DIM = 512  # bisa ubah: 50, 128, 256, 768

for idx, row in df.iterrows():
    sentence = str(row["question"])
    ids = row["id"]

    try:
        response = client.embeddings.create(
            model="text-embedding-3-small",
            input=sentence,
            encoding_format="float"
        )

        emb = response.data[0].embedding

        # cut dimensi
        emb_cut = emb[:DIM]

        # normalize
        emb_norm = normalize_l2(emb_cut)

        results.append({
            "id": int(ids),
            "text": sentence,
            "embedding": emb_norm
        })

        print(f"Processed {idx}")

        # optional: delay untuk hindari rate limit
        time.sleep(0.1)

    except Exception as e:
        print(f"Error at index {idx}: {e}")

# =========================
# SAVE JSON
# =========================
with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
    json.dump(results, f, indent=2)

print(f"\nSaved {len(results)} embeddings to {OUTPUT_JSON}")
```

```python
def load_embeddings(json_file):
    with open(json_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    texts = []
    vectors = []

    for item in data:
        texts.append(item["text"])
        vectors.append(item["embedding"])

    vectors = np.array(vectors, dtype=np.float32)

    return texts, vectors


# =========================
# USAGE
# =========================
if __name__ == "__main__":
    texts, vw = load_embeddings("OpenAIEmbeddings_256.json")

    print("Shape:", vw.shape)   # (N, 256)
    print("First vector:", vw[0])
```
